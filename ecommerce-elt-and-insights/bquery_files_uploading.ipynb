```python
import kagglehub

# Download selected version
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce/versions/7")

print("Path to dataset files:", path)
```

```python
import os

for file in os.listdir(path):
  print(file)
```

```python
import pandas as pd
from google.cloud import bigquery

#Read all files in the folder first

df_products_translation = pd.read_csv(path +"/product_category_name_translation.csv")
df_reviews = pd.read_csv(path+ "/olist_order_reviews_dataset.csv")
df_payments = pd.read_csv(path + "/olist_order_payments_dataset.csv")
df_customers = pd.read_csv(path+ "/olist_customers_dataset.csv")
df_orders = pd.read_csv(path + "/olist_orders_dataset.csv")
df_geolocation = pd.read_csv(path + "/olist_geolocation_dataset.csv")
df_sellers = pd.read_csv(path + "/olist_sellers_dataset.csv")
df_items = pd.read_csv(path + "/olist_order_items_dataset.csv")
df_products = pd.read_csv(path + "/olist_products_dataset.csv")

from google.colab import auth
auth.authenticate_user()

#Connect to the relevant bg project
client = bigquery.Client(project="project-f1e6afa5-5311-4b6e-94e")

#Upload everything to bg

job_config = bigquery.LoadJobConfig(autodetect=True, write_disposition="WRITE_TRUNCATE")

client.load_table_from_dataframe(df_products_translation, "project-f1e6afa5-5311-4b6e-94e.ecom.products_translation", job_config=job_config).result()
client.load_table_from_dataframe(df_reviews, "project-f1e6afa5-5311-4b6e-94e.ecom.reviews", job_config=job_config).result()
client.load_table_from_dataframe(df_payments, "project-f1e6afa5-5311-4b6e-94e.ecom.payments", job_config=job_config).result()
client.load_table_from_dataframe(df_customers, "project-f1e6afa5-5311-4b6e-94e.ecom.customers", job_config=job_config).result()
client.load_table_from_dataframe(df_orders, "project-f1e6afa5-5311-4b6e-94e.ecom.orders", job_config=job_config).result()
client.load_table_from_dataframe(df_geolocation, "project-f1e6afa5-5311-4b6e-94e.ecom.geolocation", job_config=job_config).result()
client.load_table_from_dataframe(df_sellers, "project-f1e6afa5-5311-4b6e-94e.ecom.sellers", job_config=job_config).result()
client.load_table_from_dataframe(df_items, "project-f1e6afa5-5311-4b6e-94e.ecom.items", job_config=job_config).result()
client.load_table_from_dataframe(df_products, "project-f1e6afa5-5311-4b6e-94e.ecom.products", job_config=job_config).result()

print("All tables uploaded!")
```
